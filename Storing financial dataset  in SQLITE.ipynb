SQLite is a software library that provides a relational database management system. The lite in SQLite means lightweight in terms of setup, database administration, and required resources.

SQLite has the following noticeable features: self-contained, serverless, zero-configuration, transactional.

Sqlite is a software library that provides us with an open-source relational data management system.

Sqlite database is just plain file

Python Sqlite3 is an excellent module with which we can perform all possible DataBase operation with in memory.


# Project

**Creating financial database in sqlite database and through this script we can download data set of SENSEX 30 companies and create the database of it.
Whenever we need the database we can directly fetch fromt sqlite database and when the script is run required data will be download and new data from last data (last date of dataset) to latest date, dataset will be appended to previous dataset automatically.**




```python
import sqlalchemy
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
wiki = "https://en.wikipedia.org/wiki/List_of_BSE_SENSEX_companies"
```

```python
#fetching script names of SENSEX
tickerSensex = pd.read_html(wiki)[0].Symbol.to_list()
```

```python
tickerSensex
```

```python
#downloading the data through yfinance
def getdata(tickers):
    data=[]
    for ticker in tickers:
        data.append(yf.download(ticker,end = "2023-01-02").reset_index())
    return data
```

```python
Sensex = getdata(tickerSensex)
```

```python
Sensex
```

```python

Indiaengine = sqlalchemy.create_engine("sqlite:///ForensicDB1.db")
    
```

```python
# storing the data into sql
def TOSQL(frames,symbols,Indiaengine):
    for frame,symbol in zip(frames,symbols):
        frame.to_sql(symbol,Indiaengine,index=False)
    print("Successfully imported data")
```

```python
TOSQL(Sensex,tickerSensex,Indiaengine)
```

```python
# downloading data from enddate + 1 to latest date and appending to previous dataset without creating duplicate of old dataset.
def sql_importer():
    for tickers in tickerSensex:
        max_date=pd.read_sql(f'SELECT MAX(Date) FROM "{tickers}"',Indiaengine).values[0][0]
        print(max_date)
        new_data = yf.download(tickers,start=pd.to_datetime(max_date))
        new_rows = new_data[new_data.index>max_date]
        new_rows.to_sql(tickers,Indiaengine,if_exists="append")
        print(str(len(new_rows))+ "new rows imported to DB")
```

```python
sql_importer()
```
